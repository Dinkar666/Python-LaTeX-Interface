```python
from sympy import *
w=sp.Symbol('w')
equa1=sp.Eq(2*w+5*x*y+2*w*y+2,0)
equa2=sp.Eq(-2*x*w+5-3*w*y,0)
display("the 2 non linear equations are:",equa1,equa2)
sol=sp.solve([equa1,equa2],x,y,dict=True)  #the solution will be stored as a dictionary
display("solution",sol)
display("1st solution set:",sp.simplify(sol[0]))
display("2nd solution set:",sp.simplify(sol[1]))
display("x value simplified from 1st set ",sp.simplify(sol[0][x]))
display("y value simplified from 1st set ",sp.simplify(sol[0][y]))
#check the validity of solutions
sol_x=sp.simplify(sol[0][x])
sol_y=sp.simplify(sol[0][y])
display("check whether the solution satify the given eqns:",sp.simplify(equa1.subs(x,sol_x).subs(y,sol_y)))
display("the x value at w=1",sol_x.subs(w,1))
display("the x value(numerical) at w=1",(sol_x.subs(w,1)).evalf(5))
display("the y value(numerical) at w=1",(sol_y.subs(w,1)).evalf(5))

display("the y value at w=1",sol_y.subs(w,1))
```

## 4.1 Derivatives
### 4.1.1 easy derivatives
- use `sp.diff("function",wrt x, number of times you want to diff)`
- use `sp.Derivative(func,variable)`: to display unsolved derivative


```python
f1=sp.Function('f')(x)
RHS=sp.exp(2*x)+sp.sin(x)
expr1= sp.Eq(f1,RHS)
display(expr1)
expr2=sp.Eq(sp.Derivative(f1),RHS.diff(x,1))
display(expr2)
expr3=sp.Eq(sp.Derivative(f1,x,2),RHS.diff(x,x))
display(expr3)
expr4=sp.Eq(sp.Derivative(f1,x,7),RHS.diff(x,7))
display(expr4)
                                          
```

### 4.1.2 Rewrite functions

```python
f=sp.Function('f')(x) # LHS of eqn
RHS1=sp.exp((sp.sinh(x)))/sp.exp(sp.exp(x)/2) #RHS of eqn
expr1=sp.Eq(f,RHS1)
RHS2=RHS1.rewrite(sp.exp) 
expr2= sp.Eq(f,RHS2) 
RHS3=RHS2.rewrite(sp.exp) 
expr3=RHS3.simplify()

expr4=sp.Eq(f,expr3)
direct=RHS1.simplify()
direct1=sp.Eq(f,direct)


display(expr1) # complete orig function printed
display("the expression in terms of exp:",expr2) #complete new eqn printed
display("simplified expression is:",expr4)  
display("try direct simplification:",direct1)  
```

### 4.1.3 Partial differentiation

```python
two_var=sp.cos(x)*sp.cot(2*y)**2
display("the 2 variable function:",two_var)
part_diff=sp.diff(two_var,y)
display("derivative wrt y",part_diff)
display("express in terms of sin",part_diff.rewrite(sp.sin))
simp=part_diff.simplify()
display("needlessly complicated: simplify first:",simp)
rewri=simp.rewrite(sp.sin)
display("simplified derivative",rewri)

display("deriv 3 times wrt x and 2 times wrt y",two_var.diff(x,x,x,y,y))
display("order doesnt matter in partial derivatives(y 2 times and x 3 times)",two_var.diff(y,2,x,3))
```
